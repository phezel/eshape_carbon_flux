## Carbon Flux Example plots 
 - Flux is positive from ocean to atmosphere 
 - Fluxes used are in dsF calculated from calc_CarbonFlux.ipynb
 - savePlots and saveFolder are defined


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import cartopy.crs as ccrs
import datetime


# read flux data from other notebook; set params for plot export/save
%store -r dsF

plt.rcParams['figure.dpi']=100
saveFolder = './example_figs/'
savePlots = False
```

## Figures 1 and 2: JJA in 2000-2009 and 2010-2019

Plot mean as color, std as size, for each gridpoint. Std size is not calibrated on plot 

```python
# FIGURE 1

dsFsub = dsF.where(dsF.time.dt.season == "JJA")
dsFsub = dsFsub.where(dsFsub.time.dt.year.isin(range(2000, 2010)))


dsMean = dsFsub[:].mean(dim='time')
dsStd = dsFsub[:].std(dim='time')
gridLat, gridLon = np.meshgrid( dsF.latitude, dsF.longitude, indexing='ij')

ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

# c is color; s is size in scatter function
# scale size as stddev
plt.scatter(gridLon, gridLat, c = dsMean[:], s = dsStd[:]*10, cmap='RdBu')

plt.title('Flux $CO_2$, SOCAT gridded data \n mean JJA, 2000-2009')
cbar = plt.colorbar()
#cbar.set_label(dsF.units)
cbar.set_label('Flux $CO_2$, ($mmol\: m^{-2}\: d^{-1}$)')

plt.clim(-20, 20)

# SAVE PLOT statements 
if savePlots : 
    plt.savefig(saveFolder +'eshapeDemo_fig_1_SOCAT_JJA_2000_2009_NAtl.png', transparent = False, facecolor = 'White')
    plt.savefig(saveFolder +'eshapeDemo_fig_1_SOCAT_JJA_2000_2009_NAtl.eps', transparent = False, facecolor = 'White')

```

```python
# FIGURE 2 
dsFsub = dsF.where(dsF.time.dt.season == "JJA")
dsFsub = dsFsub.where(dsFsub.time.dt.year.isin(range(2010, 2020)))


dsMean = dsFsub[:].mean(dim='time')
dsStd = dsFsub[:].std(dim='time')
gridLat, gridLon = np.meshgrid( dsF.latitude, dsF.longitude, indexing='ij')

ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

plt.scatter(gridLon, gridLat, c = dsMean[:], s = dsStd[:]*10, cmap='RdBu')

plt.title('Flux $CO_2$, SOCAT gridded data \n mean JJA, 2010-2019')
cbar = plt.colorbar()
#cbar.set_label(dsF.units)
cbar.set_label('Flux $CO_2$, ($mmol\: m^{-2}\: d^{-1}$)')

plt.clim(-20, 20)

if savePlots: 
    plt.savefig(saveFolder +'eshapeDemo_fig_2_SOCAT_JJA_2010_2019_NAtl.png', transparent = False, facecolor = 'White')
    plt.savefig(saveFolder +'eshapeDemo_fig_2_SOCAT_JJA_2010_2019_NAtl.eps', transparent = False, facecolor = 'White')

```

# Figure 3 - Timeseries at PAP-SO 

Plot a time series;  time series must be in the selected region already, so at minimum should select a box inclusive of the time series location.  



```python
# FIGURE 3 - TIMESERIES at PAP-SO 

# Plot a time series;  time series must be in the selected region already, so at minimum should select a box inclusive of the time series location.  

import helper_functions as hf
b = hf.get_buoy_locations()

thisBuoy = 'PAP-SO'
iLat = hf.get_nearest_index( b[thisBuoy][0] , dsF.latitude)
iLon = hf.get_nearest_index( b[thisBuoy][1] , dsF.longitude)
thisBuoyIdx = (iLat, iLon)

BOXAROUND = np.array([0])   # set as '[0]' to get only given gridpoint
#BOXAROUND = np.array([-1,0,1])  # set as [-1,0,1] to get 3x3 gridpoints around '0' as given gridpoint

fig, ax = plt.subplots()
for ii in BOXAROUND:
    for jj in BOXAROUND: 
        ax.plot(dsF.time, dsF[:, int(iLat+ii), int(iLon+jj)], 'o', markersize=2)
#         ax.plot(dsF.time, dsF.sel(latitude=slice(67.5, 69.5), longitude=slice(-10.5, -14.5)).squeeze(), '-o')

ax.set_title('Flux $CO_2$ '+ thisBuoy+ ' ('+str(b[thisBuoy][0])+'N '+str(b[thisBuoy][1])+'E) '+ ' monthly means')
ax.set_ylabel('Flux $CO_2$, ($mmol\: m^{-2}\: d^{-1}$)')

ax.set_xlim(datetime.date(2000,1,1), datetime.date(2020,1,1))

if savePlots: 
    plt.savefig(saveFolder +'eshapeDemo_fig_3_SOCAT_PAP-SO_2000-2020_timeseries.png', transparent = False, facecolor = 'White')
    plt.savefig(saveFolder +'eshapeDemo_fig_3_SOCAT_PAP-SO_2000-2020_timeseries.eps', transparent = False, facecolor = 'White')


```

# Figure 4  - Time series in ENGLISH CHANNEL 
  based on points with high number of monthly observations in gridded dataset 


```python
# FIGURE 4  - TIMESERIES in ENGLISH CHANNEL 
# --- based on points with high number of monthly observations in gridded dataset 

fig, ax = plt.subplots()

# first find latLons where count of observations is high - here set to >180 in the entire timeseries
latLon =[]
for ii in np.array(range(len(dsF.latitude))):
    for jj in np.array(range(len(dsF.longitude))): 
        
        # plot gridpoints that meet threshold number of gridded monthly observations
        if dsF[:, ii, jj].count() >=180 : 
            ax.plot(dsF.time, dsF[:, ii, jj], '-o', markersize=1, linewidth=0.25)
            latLon.append([float(dsF.latitude[ii]), float(dsF.longitude[jj])])
            
            #  polyfit a line, using time as the axis;  does not converge so turned off.  (from nan's?)  
            if len(latLon) < 0:
                coef = np.polyfit(dsF.time.astype(float), dsF[:, ii, jj], 1)
                poly1d_fn = np.polyid(coef)
                ax.plot(dsF.time, poly1d_fn(dsF.time.astype(float)), '--k')

ax.set_ylim(-15, 15)
ax.set_xlim(datetime.date(2000,1,1), datetime.date(2021,1,1))
ax.set_title('Flux $CO_2$ in English Channel')
ax.set_ylabel('Flux $CO_2$, ($mmol\: m^{-2}\: d^{-1}$)')

if savePlots: 
    plt.savefig(saveFolder +'eshapeDemo_fig_4_SOCAT_EnglishChannel_2000-2020_timeseries.png', transparent = False, facecolor = 'White')
    plt.savefig(saveFolder +'eshapeDemo_fig_4_SOCAT_EnglishChannel_2000-2020_timeseries.eps', transparent = False, facecolor = 'White')

print('Gridpoints included, (N / E):')
latLon
```
